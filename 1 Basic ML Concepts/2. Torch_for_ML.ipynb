```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

## Example of Arrays

```python
position = torch.arange(0, 60)
print("Array shape:", position.shape)
print("Array:\n", position)
```

```python
# creating a random 4x6 matrix
w = torch.randn(4, 6)
print("Array shape:", w.shape)
print("Array:\n", w)
```

### Plotting sin, cos and tan functions

```python
sin_array = torch.sin(position)
cosine_array = torch.cos(position)
tan_array = torch.tan(position)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].plot(sin_array)
axs[0].legend(['sine wave'])
axs[1].plot(cosine_array)
axs[1].legend(['cosine wave'])
axs[2].plot(tan_array)
axs[2].legend(['tan wave'])
plt.show()
```

## Dot product
Dot product is a fundamental operation in linear algebra and machine learning. It combines two vectors to produce a single scalar value, which can be interpreted in various ways depending on the context. In machine learning, the dot product is often used in algorithms such as linear regression, neural networks, and support vector machines.

Example of dot product calculation using PyTorch:
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

dot_product = 1 * 4 + 2 * 5 + 3 * 6 = 32

```python
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

print(f"Array a: {a} with shape {a.shape}")
print(f"Array b: {b} with shape {b.shape}")

dot_product = torch.dot(a, b)

print(f"\nDot product: {dot_product.item()}")
```

## Simple Neural Network

```python
# example neural network layer operation
inputs = torch.randn(150, 4)    # 150 samples, each with 4 features
weight_1 = torch.randn(4, 86)   # weights for layer 1 (4 inputs to 86 neurons)
weight_2 = torch.randn(86, 3)   # weights for layer 2 (86 inputs to 3 outputs)
bias_1 = torch.randn(86)      # bias for layer 1
bias_2 = torch.randn(3)       # bias for layer 2

# forward pass through the network
layer_1 = torch.matmul(inputs, weight_1) + bias_1  # layer 1 output
layer_2 = torch.matmul(layer_1, weight_2) + bias_2  # layer 2 output
print("Layer 1 output shape:", layer_1.shape)
print("Layer 2 output shape:", layer_2.shape)
# Note: The output of layer_2 has shape (150, 3), corresponding to 150 samples and 3 output classes.

# To get the predicted class for each sample, we can use argmax
# dim=1 to get the index of the max value in each row
classes = torch.argmax(layer_2, dim=1)
print("\nPredicted classes shape:", classes.shape)
print("Predicted classes:", classes)
```
