## Image encrypting

```python
img1 = cv2.imread('cat.jpg', 0)
img2 = cv2.imread('text.jfif', 0)

img1 = cv2.resize(img1, (600, 340))
img2 = cv2.resize(img2, (600, 340))

img1 = cv2.normalize(img1, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)
img2 = cv2.normalize(img2, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 12))
ax1.imshow(img1, cmap='gray')
ax2.imshow(img2, cmap='gray')
plt.show()
```

```python
encrypted = cv2.bitwise_xor(img1, img2)
plt.imshow(encrypted, cmap='gray');
```

```python
decrypted1 = cv2.bitwise_xor(encrypted, img2)
decrypted2 = cv2.bitwise_xor(encrypted, img1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 12))
ax1.imshow(decrypted1, cmap='gray')
ax2.imshow(decrypted2, cmap='gray')
plt.show()
```
